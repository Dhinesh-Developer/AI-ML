# Sort the dataframe

```python
# sort by one column 
# by default ascending
df_exams.sort_values(by='math score')
```

```python
# sort in descending by one column
df_exams.sort_values('math score',ascending = False)
```

```python
# sort descending my multiple column
df_exams.sort_values(['math score','reading score'],ascending = False)
```

```python
# sort descending my multiple column and update dataframe
df_exams.sort_values(['math score','reading score'],ascending = False,inplace = True)
```

```python
df_exams
```

```python
# sort with text using descending a key function
# lambda return_obj : return_obj+operations
df_exams.sort_values('race/ethnicity',ascending=True,key= lambda col: col.str.lower())
```

# The set_index() and sort_index() methods 

# create index 

```python
import numpy as np
import random
```

```python
# creating non repetitive values for the index
new_index = np.arange(0, 1000)
```

```python
random.shuffle(new_index)
```

```python
new_index
```

```python
# create a new column with new index
df_exams['new_index'] = new_index
```

```python
df_exams
```

#  set index

```python
# set new_index column as index
df_exams.set_index('new_index',inplace = True)
```

```python
df_exams
```

# Sort index

```python
# sort dataframe by new index
df_exams.sort_index()
```

```python
# sort descending using index
df_exams.sort_values(by ='new_index',ascending=False)
```

# Rename columns and index with rename()

```python
# rename column and overwrite values of the dataframe
df_exams = df_exams.rename(columns={'gender':'Gender'})
```

```python
df_exams
```

```python
# rename multiple columns and update the dataframe with inplace argument
df_exams.rename(columns={'math score':'MS',
                        'reading score':'RS',
                        'writing score':'Ws'},inplace=True)
```

```python
df_exams
```
