```python
import pandas as pd
```

```python
df = pd.read_csv('laptop_price.csv',encoding = 'latin1')
```

```python
df.head(5)
```

# 1 drop duplicates()

```python
# drop duplicates in 2 or more columns
df.drop_duplicates(['Company'])[['Company','Price_euros']]
```

```python
# sort dataframe ascending by company and price
# cheapest first and most expensive last
df = df.sort_values(['Company','Price_euros'])
df
```

```python
# cheapest: keep = 'first'
df.drop_duplicates(['Company'], keep = 'first')
```

```python
# most expensive: keep = 'last'
df.drop_duplicates(['Company'], keep = 'last',inplace= True)
df
```

```python
# Arguments
#inplace : drop duplicates in place or to return a copy.
# ignore_index = True : the resulting axis will be labled 0, 1, ...
df.drop_duplicates(['Company'], keep = 'last',inplace= True, ignore_index = True)
df
```
