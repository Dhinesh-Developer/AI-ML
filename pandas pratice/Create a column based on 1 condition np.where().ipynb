# 1. create a column based on 1 conditions: np.where()

```python
import numpy as np
```

```python
# create an array based on price tiers

new_col = np.where(df_laptops['Price_euros'] > 2000, 'Expensive', 'Cheap')
```

```python
# set it to a new column
df_laptops['Price_tier'] = new_col 
```

```python
# show first 5 columns
df_laptops.head(5)
```

```python
#count values in price_tier column
df_laptops.value_counts('Price_tier')
```
