# Add new column using the assign() and insert()

### 1.1 assign()
when to use assign?
- Add multiple columns in a single line of code
- when you need to overwrite the values of an existing columns
- it returns a new object(a copy) with the original columns in addition to new ones

```python
import numpy as np
```

```python
# create random numbers for our new score columns
score1 = np.random.randint(1,100,size = 1000)
score2 = np.random.randint(1,100,size = 1000)
```

```python
# create a series using the random numbers
series1 = pd.Series(score1, index=np.arange(0, 1000))
series2 = pd.Series(score1, index=np.arange(0, 1000))
```

```python
# using the assign() to add multiple columns
df_exams = df_exams.assign(score1=series1, score2=series2)
```

```python
df_exams
```

## 1.2 insert()
- inserts a new column at a specific position or index

```python
# using index() to add a column at a specific position
df_exams.insert(1,'test',series1)
```
