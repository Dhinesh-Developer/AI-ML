# Differnces

- loc                        iloc           select with a:
- df.loc['zero']             df.iloc[0]      value
- df.loc[['zero','one']]   df.iloc[[0,2]]    List
- df.loc[['zero':'two']]     df.iloc[0:2]    Slicing
-        both included                 0-> included 1-> excluded


# First Look at the dataset Setting index and Selecting columns

```python
import pandas as pd
```

```python
df = pd.read_csv('players_20.csv')
```

```python
df
```

```python
df.set_index('short_name', inplace = True)
```

```python
df
```

```python
#  long_name  age dob	height_cm	weight_kg	nationality	club
df = df[['long_name','age','dob','height_cm','weight_kg','nationality' ,'club']]
```

```python
df
```

# Selecting With a Single Value
- loc[row_label, column_label]

```python
# get all about L.Messi
df.loc['L. Messi']
```

```python
# get the height of L.Messi
# 2 parameter is column name
df.loc['L. Messi', 'height_cm']
```

```python
# get the height of Cristinano Ronaldo
df.loc['Cristiano Ronaldo', 'height_cm']
```

```python
# get the club of Cristinano Ronaldo
df.loc['Cristiano Ronaldo', 'club']
```

```python
# get all rows inside the 'height_cm' column
df.loc[:,'height_cm']
```

```python
# get all columns that correspond to the index 'L.Messi'
df.loc['L. Messi',:]
```

```python
df.loc['L. Messi','dob':'nationality']
```

# Selecting with a list of values

```python
# get all data about L.Messi and Cristiano Ronaldo
df.loc[['L. Messi','Cristiano Ronaldo']]
```

```python
# Get the height of L.Messi and Cristinano Ronaldo
# row , column
df.loc[['L. Messi','Cristiano Ronaldo'], 'height_cm']
```

```python
# Get the height and weight of L.Messi
df.loc[['L. Messi'], ['height_cm','weight_kg']]
```

```python
# Get Height and Weight of L.Messi and Cristiano Ronaldo
df.loc[['L. Messi','Cristiano Ronaldo'], ['height_cm','weight_kg']]
```

# 3 Selecting a range of data with a slice
- start:stop:step (Note that country to usual python slices, Both the start and the stop are included)

```python
# slicing column lables
players = ['L. Messi','Cristiano Ronaldo']
df.loc[players,'age':'club']
```

```python
df.loc[players, 'dob':'weight_kg']
```

```python
# slice index labels
columns = ['age','dob','height_cm','weight_kg']

# get top1 and top10 player name
df.index[:10]

df.loc['L. Messi':'M. Salah', columns]
```

# Selecting with conditions

```python
# one condtions : select player with height above 180cm
columns = ['age','dob','height_cm','weight_kg']

df.loc[df['height_cm'] > 180, columns]
```

```python
# multiple conditions: Select player with height above 180cm from Argentina
df[(df['nationality'] == 'Argentina') & (df['height_cm'] > 180)]
```

```python
df.loc[(df['nationality'] == 'Argentina') & (df['height_cm'] > 180), columns]
```

#  Selecting elements by index position with .iloc()

### selection with a single value
- iloc[row_position, column_position]

```python
# get the height of L. Messi
df.iloc[0,3]
```

```python
# get the weight of the crisitiano Ronaldo
df.iloc[1, 4]
```

```python
# get all rows inside the 'height_cm' column
df.iloc[:,3]
```

```python
# get all columns that correspond to the index 'L. Messi'
df.iloc[0, :]
```

## select elements with a list of values

```python
# get all data about L. Messi and Cristiano Ronaldo
df.iloc[[0,2]]
```

```python
# Get the height of L.Messi and Ciristiano Ronaldo
df.iloc[[0, 1], 3]
```

```python
# get the height and weight of L.Messi
df.iloc[[0,1], [3]]
```

```python
# get height and weight of L.Messi and Cristiano Ronaldo
df.iloc[[0,1],[3,4]]
```
