# 2. Read .csv from multiple URL's with pandas
https://www.football-data.co.uk/mmz4281/2122/E0.csv
- Link: root+season+league

```python
# link structure
"https://www.football-data.co.uk/mmz4281/"+"2122"+"/"+"E0"+".csv"
```

```python
# creat a root variable
root = "https://www.football-data.co.uk/mmz4281/"
```

```python
# create a list of leagues
leagues = ['E0','E1','E2','E3','EC']
frames = []

#looping through leagues, read multiple csv and append it into a list
for league in leagues:
    df = pd.read_csv(root + '2122' + '/' + league + '.csv', encoding = 'unicode_escape')
    frames.append(df)
```

```python
# length of the frame
len(frames)
```

```python
# show 1sr, 2nd and 3rd element
frames[0]
```

```python
frames[4]
```

# Multiple Seasons

```python
for season in range(15,21):
    print(str(season)+str(season+1))
```

```python
# create a list of leagues
leagues = ['E0','E2','E3']
frames = []

#looping through leagues, read multiple csv and append it into a list
for league in leagues:
    for season in range(15,21):
        df = pd.read_csv(root + str(season) + str(season+1) + '/' + league + '.csv')
        df.insert(1, 'season', season)
        frames.append(df)
```

```python
# total frames (6 seasons x 3 leagues)
len(frames)
```

# Organize All the Data in a Dictionary

```python
# create a dictionary with original name of league as key
dict_countries = {
    'Spanish Ls Liga':'SP1',
    'Spanish Segunda Division':'SP2',
    'German Buneslinga':'D1',
    'English Premier League':'E0',
    'English League 1':'E2',
    'English League 1':'E3'
}
```

```python
dict_countries['Spanish Ls Liga']
```

```python
for league in dict_countries:
    print(dict_countries)
    print(dict_countries[league])
```

```python
dict_historical_data = {}

#looping t
# create a list of leagues
# leagues = ['E0','E2','E3']


#looping through leagues, read multiple csv and append it into a list
for league in dict_countries:
    frames = []
    for season in range(15,21):
        df = pd.read_csv(root + str(season) + str(season+1) + '/' + dict_countries[league] + '.csv')
        df.insert(1, 'season', season)
        frames.append(df)
    df_concat = pd.concat(frames)  
    dict_historical_data[league] = df_concat
```
